```python
#Kütüphaneleri ve datasetini indirelim.
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('spotify_data.csv' )
df = data.copy()
```

```python
#Datasetini inceleyelim.
print(df.info())
```

```python
print('Kolonların tipleri:',df.dtypes)
print('-'*80)
print('Kolonlar:',df.columns)
print('-'*80)
print('Sütun sayısı:',df.shape[0])
print('Kolon sayısı:',df.shape[1])
```

```python
df.head()
```

```python
#Çok fazla kolon var bazılarını siliyorum
df.drop(['danceability', 'loudness','loudness', 'acousticness','instrumentalness', 'valence','energy','speechiness','liveness','tempo','duration_ms'], axis=1, inplace=True)
```

```python
df
```

```python
#Boş değer olan satırları görelim ve silelim.
df.isnull().sum()
```

```python
df.dropna()
```

```python
#Müzik türlerine bakalım
list=[]
türler=df["genre"].unique()
for tür in türler:
    list.append(tür)
sayı=df["genre"].nunique()
print(f'{sayı}adet müzik türü vardır .\n{türler}')
```

```python
#Tango müzik türündeki ilk 50 şarkıyı görelim
df[df["genre"] =='tango'].head(50)
```

```python
#Verimizi müzik türlerine göre gruplayıp müzik türlerini popülerliğe göre sıralayalım.
genres = df.groupby('genre')
```

```python
genres["popularity"].mean().sort_values(ascending=False)
```

```python
#2020 yılındaki en pöpüler şarkıların bilgilerine bakalım.
df[df['year']==2020].sort_values('popularity',ascending=False)[0:20]
```

```python
#En popüler şarkı
maxpopularity= df['popularity'].max()
enpopülerşarkı= df[df['popularity'] == maxpopularity]
print(f'En popüler şarkı: {enpopülerşarkı["artist_name"].iloc[0]}-{enpopülerşarkı["track_name"].iloc[0]}')
```

```python
#En çok sarkısı olan 20 sanatçı
top20 = df['artist_name'].value_counts().head(20)
print(top20)
```

```python
#En fazla şarkısı olan 5 sanatçıyı ve şarkı sayılarının kendi aralarındaki yüzdelerini pasta grafiğinde görelim
top5 = df['artist_name'].value_counts().head(5)
plt.figure(figsize=(13,7))
plt.pie(top5, labels=top5.index, autopct='%1.0f%%')
plt.title('En Çok Şarkısı Olan 5 Sanatçı')
plt.show()
```

```python
#Hangi türde ne kadar şarkı olduğunu gösteren grafik
genre_counts = df['genre'].value_counts()
plt.figure(figsize=(15,6))
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Counts')
plt.xticks(rotation=90)
plt.show()
```
